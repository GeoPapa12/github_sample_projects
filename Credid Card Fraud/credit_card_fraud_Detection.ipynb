# Credit Card Fraud Detection

## Goal for this Notebook:

The aim of the project is to predict fraudulent credit card transactions using machine learning models. This is crucial from the bank’s as well as customer’s perspective. The banks cannot afford to lose their customers’ money to fraudsters. Every fraud is a loss to the bank as the bank is responsible for the fraud transactions.

The dataset contains transactions made over a period of two days in September 2013 by European credit cardholders. The dataset is highly unbalanced. We need to take care of the data imbalance while building the model and come up with the best model by trying various algorithms.

Before using an oversmapling method like SMOTE, it is very important to split the train and test datasets before oversampling the dataset (the train dataset)

```python
# Case Oversampling
# ===============================================================
# balancing the data by Undesampling of the dominant class + Oversamping of the minority class

# Spliting the dataset before oversampling with SMOTE
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline

X = df.drop(['Class'], 1)
y = df['Class']
print("df len: ", len(df))
#===============================================================================================
df1 = df[df.Class == 1]
df0 = df[df.Class == 0]

df1_test = df1.iloc[0:200]
df0_test = df0.iloc[0:200]
df_test = pd.concat([df1_test, df0_test])
print('Test Class 0: ', len(df_test[df_test['Class']==0]))
print('Test Class 1: ', len(df_test[df_test['Class']==1]))
print("df Test len: ", len(df_test))

df1_train = df1.iloc[200:-1]
df0_train = df0.iloc[200:-1]
df_train = pd.concat([df1_train, df0_train])
print('Train Class 0: ', len(df_train[df_train['Class']==0]))
print('Train Class 1: ', len(df_train[df_train['Class']==1]))
print("df Train len: ", len(df_train))

X_train = df_train.drop(['Class'], 1)
y_train = df_train['Class']

X_test = df_test.drop(['Class'], 1)
y_test = df_test['Class']

#===============================================================================================
over = SMOTE(sampling_strategy=0.25)
under = RandomUnderSampler(sampling_strategy=1)
steps = [('o', over), ('u', under)]
pipeline = Pipeline(steps=steps)
#===============================================================================================

X_train, y_train = pipeline.fit_resample(X_train, y_train)

df_over = X_train.copy()
df_over['Class'] = y_train
print('SMOTE Class 0: ', len(df_over[df_over['Class']==0]))
print('SMOTE Class 1: ', len(df_over[df_over['Class']==1]))
print("df SMOTE len: ", len(df_over))

results_SMOTE = ML.ML_Basic_Models(X_train, y_train, X_ex_test=X_test, y_ex_test=y_test)
```
